```python
# Organize data into train, valid, test dirs
os.chdir('D:\work\ML_project\cat or dogs model\Sign-Language-Digits-Dataset-master\Dataset')
if os.path.isdir('train/0/') is False: 
    os.mkdir('train')
    os.mkdir('valid')
    os.mkdir('test')

    for i in range(0, 10):
        shutil.move(f'{i}', 'train')
        os.mkdir(f'valid/{i}')
        os.mkdir(f'test/{i}')

        valid_samples = random.sample(os.listdir(f'train/{i}'), 30)
        for j in valid_samples:
            shutil.move(f'train/{i}/{j}', f'valid/{i}')

        test_samples = random.sample(os.listdir(f'train/{i}'), 5)
        for k in test_samples:
            shutil.move(f'train/{i}/{k}', f'test/{i}')
os.chdir('../..')
```

```python
train_path = 'Sign-Language-Digits-Dataset-master/Dataset/train'
valid_path = 'Sign-Language-Digits-Dataset-master/Dataset/valid'
test_path = 'Sign-Language-Digits-Dataset-master/Dataset/test'

train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input).flow_from_directory(
    directory=train_path, target_size=(224,224), batch_size=10)
valid_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input).flow_from_directory(
    directory=valid_path, target_size=(224,224), batch_size=10)
test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input).flow_from_directory(
    directory=test_path, target_size=(224,224), batch_size=10, shuffle=False)
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
mobile.summary()
```

```python


base_model_layers = mobile.layers[:-5]
model = Sequential()
for layer in base_model_layers:
    model.add(layer)
model.add(Reshape(target_shape=(7*7*1024,)))
model.add(Dense(units=10, activation='softmax'))
```

```python
model.summary()
```

```python
# x = mobile.layers[-5].output

# x = tf.keras.layers.Reshape(target_shape=(1024,))(x)
# output = Dense(units=10, activation='softmax')(x)
# model = Model(inputs=mobile.input, outputs=output)
```

```python
for layer in model.layers[:-22]:
    layer.trainable = False
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=train_batches,
            steps_per_epoch=len(train_batches),
            validation_data=valid_batches,
            validation_steps=len(valid_batches),
            epochs=10,
            verbose=2
)
```

```python
loss, accuracy = model.evaluate(test_batches)

print(f"The Accuracy Testing Data is ==> {accuracy} %")
print(f"The loss Testing Data is ==> {loss}")
```
